#### **Importing Required Libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
```

#### **Reading the Image**

```python
image1=cv2.imread('/content/sample_data/elonmusk image.jpg')
cv2_imshow(image1)
```

#### **Converting the Original image to Grayscale**

```python
gray_image=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

#### **Inverting the grayscale image**

```python
inverted_gray=cv2.bitwise_not(gray_image)
cv2_imshow(inverted_gray)
```

#### **Applying Gaussian Blur**

```python
blurred_inverted_image=cv2.GaussianBlur(inverted_gray,(21,21),sigmaX=0,sigmaY=0)
cv2_imshow(blurred_inverted_image)
```

#### **Pencil Sketch**

```python
def pencil_sketch(x,y):
  return cv2.divide(x,255-y,scale=256)

pencil_sketch_image=pencil_sketch(gray_image,blurred_inverted_image)
cv2_imshow(pencil_sketch_image)
```

**Displaying Original Image and Pencil Sketch Image**

```python
cv2_imshow(image1)
cv2_imshow(pencil_sketch_image)
```
